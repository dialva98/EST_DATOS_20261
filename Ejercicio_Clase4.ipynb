```python
#Hacer un código que permita crear en el disco duro un archivo con una matriz de 100.000x100.000 con ceros y unos como entradas de la matriz
import numpy as np

def archivo(n,file="matriz.txt"): #tamaño de la matriz nxn y nomenclatura del archivo de salida
    with open(file,"w") as f: #manipulacion de file en una variable temporal
        for i in range(n):
            fila = ["0"] * n #vector fila 1xn (con ceros)
            fila[i] = "1" #entrada diagonal
            f.write(" ".join(fila) + "\n") #apilar por fila y crear matriz
    return file
```

```python
archivo(100000)
```
